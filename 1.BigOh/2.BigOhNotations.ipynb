#### Simplified Analysis of an Algorithms Effeciency


###### We can give performance measure of an algorithm using `Big Oh` Notations 

    1. Complexity in terms of input size N
    2. Machine Independent 
    3. Basic Compute steps (key operations)
    4. Time & Space Complexity
    

```python
names = [ 'yadvendra', 'rajat', 'kushal', 'mayank', 'ravi', 'simran', 'gaurav', 'akhilesh', 'vijay', 'sachin' ] 

def search(name):
    flag = False # 1
    for user in names: # N
        if user == name:  
            result = "Found" 
            flag = True
    if flag == False: # 1
        result = "Not Found" # 1
    
    return result # 1


# Best == Worst O(n)
```

```python
names = [ 'yadvendra', 'rajat', 'kushal', 'mayank', 'ravi', 'simran', 'gaurav', 'akhilesh', 'vijay', 'sachin' ] # O(1)

def search(name):
    for user in names: # len(n), N
        if user == names:  # O(N)
            print("Found") # O(1)
            break # O(1)
    else:
        print("Not Found")
        
search('yadvendra')
# Best Case Complexity 
# T(n) = Omega(4) => Omega(1) -> Constant
search('sachin')
search('salman khan')
# Worst Case Complexity
# T(n) =O(1 + N + 1 + 1) => O(N + 3) => O(N) # Linear Complexity


# Average Case Complexity
search('simran')
# T(n) = Theta(n/2) = Theta(n)
```

#### Type of Measurement

1. Worst Case - Big Oh Notation $\mathcal{O}(n\log n)$
    
        maximum time taken by a program to solve a problem
        
2. Best Case  - Big Omega Notation $\Omega (n\log n)$

        minimum time taken by a program to solve a problem
        
3. Average Case Complexity - Theta Notation $\theta (n\log(n))$

        on average time taken by a program to solve a problem

### Rules to write Big Oh notations 
    
    1. ignore constant
        
        T(n) = O(400) + O(6000) + O(10) + O(N) ====> O(N) 
        
    2. Certain terms "Dominate" Other Terms 
    
        remove all non dominate terms only write dominate terms in BigOh notation
        
        O(1) < O(logn) < O(n) < O(nlogn) < O(n^2) < O(2^n) < O(n!)
        
        
        T(n) = O(500) + O(n) + O(nlog(n)) + O(n^2) ===> O(n^2)
        
    3. Use Different Variables for Different input sizes
    
        T(n) = O(n) + O(m) + O(100) => O(n+m)
        
        if n > m => T(n) = O(n) else T(n) = O(m)
